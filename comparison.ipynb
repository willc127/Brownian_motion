```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('rdiff.csv')
data_box = pd.read_csv('rdiff_box.csv')
data_force = pd.read_csv('rdiff_force.csv')
```

```python
MSD = data['MSD']
tau = data['tau']
MSD_box = data_box['MSD']
tau_box = data_box['tau']
MSD_force = data_force['MSD']
tau_force = data_force['tau']
```

```python
fig1 = plt.figure(figsize=(8,6))
plt.plot(tau_force, MSD_force, '-r',label='Force (superdiffusion)')
plt.plot(tau, MSD, '-b',label='Free particle (normal diffusion)')
plt.plot(tau_box, MSD_box, '-g',label='Inside a box (subdiffusion)')
plt.legend(loc='best', frameon=False)
plt.xlabel('Time lag')
plt.ylabel('MSD')
```
